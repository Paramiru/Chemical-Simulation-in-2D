```python
from pandas import read_csv
from seaborn import set_theme
import matplotlib.pyplot as plt
set_theme()
```

```python
data = read_csv('../results/fraction_of_points_question_b.csv')
times, frac_a, frac_b, frac_c = data['time'], data['fraction_a'], data['fraction_b'], data['fraction_c']
```

```python
plt.figure(figsize=(14, 10))
plt.plot(times, frac_a, label='type a')
plt.plot(times, frac_b, label='type b')
plt.plot(times, frac_c, label='type c')
plt.xlabel('Time')
plt.ylabel('Average fraction of cells of a given type')
plt.legend()
```
