```python
import numpy as np
from pandas import read_csv
import matplotlib.pyplot as plt
```

```python
data = read_csv('../results/two_points_data_question_e.csv')
```

```python
times, value1s, value2s = data['time'].to_numpy(), data['value1'].to_numpy(), data['value2'].to_numpy()
```

```python
plt.figure(figsize=(14,10))
plt.plot(times, value1s)
plt.plot(times, value2s)

plt.xticks(list(range(0, times[-1], 250)))
plt.show()
```
