```python
from pandas import read_csv
from seaborn import set_theme
import matplotlib.pyplot as plt
set_theme()
```

```python
data = read_csv('../results/questionf_D_0.3.csv')
probabilities, distance = data['p'].to_numpy(), data['r'].to_numpy()
plt.figure(figsize=(12,8))
plt.scatter(distance, probabilities)
plt.xlabel('distance between two cells r')
plt.ylabel('Probability of two cells with same type')
```
